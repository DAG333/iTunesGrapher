# iTunesGrapher

This script is designed to analyze and graph your iTunes library through the metadata that is stored within the native iTunes xml file. To run this script, first, you need to import the package.

```python
import iTunesGrapher as ig
```

To better analyze the xml, this function creates a SQLite database for easy data storage, retrieval and organization. 

```python
ig.write_database("iTunes_library.xml","library.db")
```

Once you have created the database file, you can run analysis functions to gather interesting information about your iTunes library and person music listening habits. 

```python
%matplotlib inline

ig.graph_library_summary("library.db")
```

```python
ig.graph_genres("library.db")
```
